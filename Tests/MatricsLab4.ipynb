### Setup

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

x = np.array([i for i in range(1, 50000*9+1, 50000)])

yEnQueue = np.array([ 0, 0.5, 1, 2, 2, 3, 4, 5, 6])

yDeQueue = np.array([ 0, 0.3, 1, 1, 2, 2, 2, 3, 4])
```

### Enqueue

```python
xnew = np.linspace(x.min(), x.max(), 500)
spline = make_interp_spline(x, yEnQueue)

x_coords = np.array([0, 50000])
y_coords = np.array([0, 1])

# Calculate the coefficients of the linear function
coefficients = np.polyfit(x_coords, y_coords, 1)

# Generate x values
x_values = np.linspace(0, x.max(), 100)

# Calculate the corresponding y values
y_values = coefficients[0]*x_values + coefficients[1]
plt.plot(x_values, y_values, 'b-', label = 'teor')
plt.plot(xnew, spline(xnew), 'r-', x, yEnQueue, 'bo', label='exp')
plt.legend()
plt.show()
```

### Dequeue

```python
xnew = np.linspace(x.min(), x.max(), 500)
spline = make_interp_spline(x, yDeQueue)

x_coords = np.array([0, 50000])
y_coords = np.array([0, 1])

# Calculate the coefficients of the linear function
coefficients = np.polyfit(x_coords, y_coords, 1)

# Generate x values
x_values = np.linspace(0, x.max(), 100)

# Calculate the corresponding y values
y_values = coefficients[0]*x_values + coefficients[1]
plt.plot(x_values, y_values, 'b-', label = 'teor')
plt.plot(xnew, spline(xnew), 'r-', x, yDeQueue, 'bo', label='exp')
plt.legend()
plt.show()
```
